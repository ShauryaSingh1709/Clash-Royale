```python

DATA_PATH = "../dataset/raw/"


cards_df = pd.read_csv(DATA_PATH + "cards.csv")
decks_df = pd.read_csv(DATA_PATH + "decks.csv")
battles_df = pd.read_csv(DATA_PATH + "battles.csv")
meta_df = pd.read_csv(DATA_PATH + "meta_stats.csv")

print("All datasets loaded successfully!\n")
print(f"Cards Dataset Shape:    {cards_df.shape}")
print(f"Decks Dataset Shape:    {decks_df.shape}")
print(f"Battles Dataset Shape:  {battles_df.shape}")
print(f"Meta Stats Shape:       {meta_df.shape}")
```

```python

print("MISSING VALUES ANALYSIS")
print("=" * 50)

datasets = {
    "Cards":    cards_df,
    "Decks":    decks_df,
    "Battles":  battles_df,
    "Meta":     meta_df
}

for name, df in datasets.items():
    missing = df.isnull().sum().sum()
    print(f"{name:10} → Missing values: {missing}")
```

```python

print("DUPLICATE ROWS ANALYSIS")
print("=" * 50)

for name, df in datasets.items():
    duplicates = df.duplicated().sum()
    print(f"{name:10} → Duplicate rows: {duplicates}")
```

```python

print("UNIQUE VALUES IN CARDS DATASET")
print("=" * 50)

print(f"\nTotal unique cards: {cards_df['name'].nunique()}")
print(f"\nRarity distribution:")
print(cards_df['rarity'].value_counts())
print(f"\nCard types:")
print(cards_df['type'].value_counts())
print(f"\nArchetypes:")
print(cards_df['archetype'].value_counts())
print(f"\nElixir cost range: {cards_df['elixir_cost'].min()} - {cards_df['elixir_cost'].max()}")
```

```python

print(" TOP 10 CARDS BY WIN RATE")
print("=" * 50)

top_cards = cards_df.nlargest(10, 'win_rate')[['name', 'rarity', 'elixir_cost', 'win_rate', 'usage_rate']]
top_cards
```

```python
# ====================================================
# 📦 IMPORT LIBRARIES
# ====================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Settings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
sns.set_style('darkgrid')

print("✅ All libraries imported successfully!")
print(f"📊 Pandas version: {pd.__version__}")
print(f"🔢 NumPy version: {np.__version__}")
```

```python
# ====================================================
# 📂 LOAD ALL DATASETS
# ====================================================

# Define dataset paths
DATA_PATH = "../dataset/raw/"

# Load each dataset
cards_df = pd.read_csv(DATA_PATH + "cards.csv")
decks_df = pd.read_csv(DATA_PATH + "decks.csv")
battles_df = pd.read_csv(DATA_PATH + "battles.csv")
meta_df = pd.read_csv(DATA_PATH + "meta_stats.csv")

print("✅ All datasets loaded successfully!\n")
print(f"📦 Cards Dataset Shape:    {cards_df.shape}")
print(f"🃏 Decks Dataset Shape:    {decks_df.shape}")
print(f"⚔️  Battles Dataset Shape:  {battles_df.shape}")
print(f"📊 Meta Stats Shape:       {meta_df.shape}")
```

```python
# ====================================================
# 📊 PREVIEW: META STATS DATASET
# ====================================================

print("=" * 70)
print("📊 META STATS DATASET - First 10 Rows")
print("=" * 70)
meta_df.head(10)
```

```python
# ====================================================
# 🔍 MISSING VALUES CHECK
# ====================================================

print("🔍 MISSING VALUES ANALYSIS")
print("=" * 50)

datasets = {
    "Cards":    cards_df,
    "Decks":    decks_df,
    "Battles":  battles_df,
    "Meta":     meta_df
}

for name, df in datasets.items():
    missing = df.isnull().sum().sum()
    print(f"📦 {name:10} → Missing values: {missing}")
```

```python
# ====================================================
# 🔁 DUPLICATE ROWS CHECK
# ====================================================

print("🔁 DUPLICATE ROWS ANALYSIS")
print("=" * 50)

for name, df in datasets.items():
    duplicates = df.duplicated().sum()
    print(f"📦 {name:10} → Duplicate rows: {duplicates}")
```

```python
# ====================================================
# 🎴 UNIQUE VALUES IN CARDS
# ====================================================

print("🎴 UNIQUE VALUES IN CARDS DATASET")
print("=" * 50)

print(f"\n🃏 Total unique cards: {cards_df['name'].nunique()}")
print(f"\n⭐ Rarity distribution:")
print(cards_df['rarity'].value_counts())
print(f"\n🎯 Card types:")
print(cards_df['type'].value_counts())
print(f"\n🏆 Archetypes:")
print(cards_df['archetype'].value_counts())
print(f"\n💧 Elixir cost range: {cards_df['elixir_cost'].min()} - {cards_df['elixir_cost'].max()}")
```

```python
# ====================================================
# 🏆 TOP 10 CARDS BY WIN RATE
# ====================================================

print("🏆 TOP 10 CARDS BY WIN RATE")
print("=" * 50)

top_cards = cards_df.nlargest(10, 'win_rate')[['name', 'rarity', 'elixir_cost', 'win_rate', 'usage_rate']]
top_cards
```
